```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.columns = ['davlat','kod','yil','komir','gaz','gidro','yangilanuvchi','quyosh','neft','shamol','atom']
df.head()
```

```python
df.shape
```

```python
df['davlat'].unique()
```

```python
dfus = df[df['davlat'] == 'United States'].copy()
dfus
```

```python
dfus.set_index('yil', inplace=True)
df.set_index('yil', inplace=True)
df
```

```python
dfus
```

```python
energy = ['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']
electricity_us_2020 = dfus.loc[2020, energy].sum()
electricity_us_2020
```

```python
energy_sources = ['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']
electricity_avg_by_source = dfus.loc[2010:2020, energy_sources].mean()
electricity_avg_by_source_sorted = electricity_avg_by_source.sort_values(ascending=False)
electricity_avg_by_source_sorted
```

```python
min_hydro_year = dfus['gidro'].idxmin()
max_hydro_year = dfus['gidro'].idxmax()
min_hydro_year
```
